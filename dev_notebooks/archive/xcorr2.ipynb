# Cross Correlation Experiment 2

```python
import scipy
from scipy import signal
from scipy import io
import numpy as np
from matplotlib import pyplot as plt
```

```python
mat_dict_bb = io.loadmat('2022_12_2_1100_broadband.mat')
mat_dict_lf = io.loadmat('2022_12_2_1100_lowfrequency.mat')
```

```python
ec = mat_dict_lf['easter_caldera'] - np.mean(mat_dict_lf['easter_caldera'])
cc = mat_dict_lf['central_caldera'] - np.mean(mat_dict_lf['central_caldera'])
```

```python
b,a = signal.butter(4, [0.01,0.9], btype='bandpass')

ec_filt = signal.filtfilt(b,a,ec)
cc_filt = signal.filtfilt(b,a,cc)
```

```python
R = signal.fftconvolve(np.squeeze(cc_filt), np.flip(np.squeeze(ec_filt)))
tau = np.arange(0, 2880001/200, 1/200)
tau = tau - tau[-1]/2
```

```python
plt.plot(tau, R)
plt.xlim([-10,10])
```
