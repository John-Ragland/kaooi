```python
import ooipy
import xarray as xr
import numpy as np
from matplotlib import pyplot as plt
import os
from dask.distributed import Client
from OOI_hydrophone_cloud import utils
from OOI_hydrophone_cloud.processing import processing
import pandas as pd
import hvplot.xarray
from scipy import signal
from datetime import datetime
```

```python
client = Client()
client
```

```python
storage_options = {'account_name':'lfhydrophone', 'account_key':os.environ['AZURE_KEY_lfhydrophone']}
ds = xr.open_zarr('abfs://hydrophonedata/lf_hydrophone_data_test.zarr', storage_options=storage_options)
```

```python
start_time = pd.Timestamp('2021-06-24 21:00:00')
end_time = pd.Timestamp('2021-06-25 02:00:00')

ds_sliced = utils.slice_ds(ds, start_time, end_time)
```

```python
caldera_hydrophones = ds_sliced[['AXCC1', 'AXEC2']]
```

```python
NCCF_stack = processing.compute_NCCF_stack(caldera_hydrophones, compute=True, stack=False)
NCCF_stack.name='full stack'
NCCF_stack = NCCF_stack.assign_coords({'time':pd.to_datetime(np.linspace(start_time.value, end_time.value, 600))})
```

```python
NCCF_stack_c = xr.DataArray(np.abs(signal.hilbert(NCCF_stack.values)), dims=NCCF_stack.dims, coords=NCCF_stack.coords, name=NCCF_stack.name)
```

```python
NCCF_stack_c.hvplot.image(x='time', y='delay', cmap='inferno', rasterize=True)
```

```python
NCCF = NCCF_stack.mean('time')
```

```python
NCCF.hvplot.line()
```
