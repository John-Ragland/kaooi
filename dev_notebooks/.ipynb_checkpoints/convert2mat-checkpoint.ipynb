## Get Low Frequency Data

```python
storage_options = {'account_name':'lfhydrophone', 'account_key':os.environ['AZURE_KEY_lfhydrophone']}
ds = xr.open_zarr('abfs://hydrophonedata/lf_hydrophone_data_test.zarr', storage_options=storage_options)
```

```python
start_time = pd.Timestamp('2021-06-24 T21:00')
end_time = pd.Timestamp('2021-06-25 T02:00')
```

```python
marasondo_slice = utils.slice_ds(ds, start_time, end_time)
```

```python
# convert eastern and central caldera to .mat file
lf_data = {'central_caldera':marasondo_slice.AXCC1.values, 
           'eastern_caldera':marasondo_slice.AXEC2.values}
           #'time_coord':marasondo_slice.time.values.astype(int)}
```

```python
savemat('/datadrive/kauai/lf_hydrophones_20210624T2100_20210625T0200.mat', lf_data)
```

```python
with open('/datadrive/kauai/lf_hydrophones_20210624T2100_20210625T0200.pkl', 'wb') as f:
    pickle.dump(lf_data, f)
```

## Get Broadband Data

```python
import numpy as np
import obspy
import os
from tqdm import tqdm
```

```python
fbase = '/datadrive/kauai/mseed_files/'
fs = os.listdir('/datadrive/kauai/mseed_files/')
fns = [fbase+f for f in fs]
```

```python
pc03A = []
pc01A = []

for f in fns:
    if f[32] == 'A':
        pc03A.append(f)
    else:
        pc01A.append(f)
```

```python
pc03A_streams = []
for k, f in enumerate(tqdm(pc03A)):
    pc03A_streams.append(obspy.read(f))
```

```python
for k, x in enumerate(pc03A_streams):
    if k == 0:
        pc03A_stream = x
    else:
        pc03A_stream += x
```
