```python
import kaooi
import xarray as xr
import numpy as np
import xrsignal
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
rep = kaooi.construct_replica(sampling_rate=200)
```

```python
reph = kaooi.construct_replica(sampling_rate=3000)

fig = plt.figure(figsize=(10,2))
reph.plot()
plt.xlabel('time [s]')
plt.ylabel('amplitude')
plt.xlim([0.5,0.75])
plt.grid()

fig.savefig('figures/sequence.svg')
```

```python
Rxx = xrsignal.correlate(rep, rep, mode='same')
Rxx = Rxx.rename({'time':'delay'}).assign_coords({'delay':np.arange(-13.64, 13.64,0.005)})
```
